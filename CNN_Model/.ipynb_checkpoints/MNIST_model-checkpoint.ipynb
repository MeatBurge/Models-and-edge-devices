#### 1. Dataset
   
   ##### 1.1 Loading and analyzing the dataset

````python
# Import the required libraries 
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras.datasets import mnist
````

#### 1.2. Load the data set 

````python
# Load the data set 
(x_train, y_train), (x_test, y_test) = tf.keras.datasets.mnist.load_data()
````

#### 1.3. the training and test data.
We need to shape the data differently. Since we're treating the data as 2D images of 28x28 pixels instead of a flattened stream of 784 pixels, we need to shape it accordingly. Depending on the data format Keras is set up for, this may be 1x28x28 or 28x28x1 (the "1" indicates a single color channel, as this is just grayscale. If we were dealing with color images, it would be 3 instead of 1 since we'd have red, green, and blue color channels) 

````python
# scaling all values are in the [0, 1] interval.
x_train = x_train.astype("float32") / 255
x_test = x_test.astype("float32") / 255

# Make sure images have shape (28, 28, 1)
x_train = np.expand_dims(x_train, axis=3)
x_test = np.expand_dims(x_test, axis=3)
 
print('the train data shape:',x_train.shape)
print('the train label shape:',y_train.shape)
print('the test data shape:',x_test.shape)
print('the test label shape:',y_test.shape)
````

#### 1.4.  Convert train and test labels to be categorical in one-hot format. 

````python
print('y_train data:',y_train)
y_train_labels =  y_train

# Convert train and test labels to be categorical in one-hot format
from keras.utils import to_categorical

y_train = to_categorical(y_train, 10)
y_test = to_categorical(y_test, 10)

print('\ny_train labels with one-hot format:')
print(y_train)

print('\nthe train label shape:',y_train.shape)
print('the test label shape:',y_test.shape)
````

#### 1.5.  Let's print out one of the training images with its label. 

````python
row, col = 5, 10
fig, axes = plt.subplots(row, col, figsize=(15, 9))

# Output 25 training images and their labels.
for i in range(col):
    for j in range(row):
        axes[j][i].set_xlabel(str("Digits:{}".format(y_train_labels[i*col+j*row])), fontsize=10)
        axes[j][i].set_xticklabels([])
        axes[j][i].set_yticklabels([])
        axes[j][i].tick_params(bottom=False, left=False, top=False, right=False)
        axes[j][i].imshow(x_train[i*col+j*row], cmap='gray_r')

plt.show()
````

### 2. Model
#### 2.1.   Setting up a convolutional neural network involves more layers. 
+ a.  Start with a 2D convolution of the image - it's set up to take 32 windows, or "filters", of each image, each filter being 3x3 in size. 
+ b.  Run a second convolution on top of that with 64 3x3 windows - this topology is just what comes recommended within Keras's own examples. Again you want to re-use previous research whenever possible while tuning CNN's. 
+ c.  Next we apply a MaxPooling2D layer that takes the maximum of each 2x2 result to distill the results down into something more manageable. 
+ d.  A dropout filter is then applied to prevent overfitting. 
+ e.  Next flatten the 2D layer we have at this stage into a 1D layer.... and feed that into a hidden, flat layer of 128 units. 
+ f.  Apply dropout again to further prevent overfitting. 
+ g.  And finally, feed that into our final 10 units where softmax is applied to choose our category of 0-9. 
+ h.   Check the model description. [Hint: model.summary()] 

````python
#  function : create cnn
def create_cnn():
    cnn_model = tf.keras.models.Sequential([
        #  Convolution
        tf.keras.layers.Conv2D(filters=32, kernel_size=3, activation='relu', input_shape=(28, 28, 1)),
        # Pooling
        tf.keras.layers.MaxPooling2D(pool_size=2, strides=2),
        # second convolutional layer
        tf.keras.layers.Conv2D(filters=64, kernel_size=3, activation='relu'),
        tf.keras.layers.MaxPooling2D(pool_size=2, strides=2),
        tf.keras.layers.Dropout(0.2),
        # Flattening
        tf.keras.layers.Flatten(),
        # Full Connection
        tf.keras.layers.Dense(units=128, activation='relu'),
        tf.keras.layers.Dropout(0.2),
        # Output Layer
        tf.keras.layers.Dense(units=10, activation='softmax')
    ])
    return cnn_model


# create cnn model
model = create_cnn()
# the cnn model description
model.summary()
````

#### 2.2 Train the model and tuner parameters

````python
# define the grid search parameters
parameters = {'epochs': 10, 'batch_size': [32, 64, 128, 256]}

# search over specified parameter values for an estimator.
historys = []
for size in parameters['batch_size']:
    # create cnn model
    cnn_model = create_cnn()
    cnn_model.compile(optimizer='adam',loss='categorical_crossentropy', metrics=['accuracy'])
    history = cnn_model.fit(
        x_train, y_train, epochs=parameters['epochs'], batch_size=size, validation_data=(x_test, y_test))
    historys.append(history)
````

````python
# Plot loss and  accuracy of different batch_size and epochs
i = 0
for history in historys:
    fig,ax = plt.subplots(1,2,figsize=(16,9))

    epochs = parameters['epochs']
    x = np.linspace(1,epochs,epochs)
    # Plot out the loss and the val_loss
    axes = ax[0]
    axes.plot(x, history.history['loss'], marker='o', label='loss',)
    axes.plot(x, history.history['val_loss'], marker='o', label='val_loss')
    axes.set_title('batch_size=' + str(parameters['batch_size'][i]),fontsize=12)
    axes.set_xlabel('epochs',fontsize=12)
    axes.set_ylabel('loss',fontsize=12)
    axes.legend(fontsize=12)
    axes.grid()

    # Plot out the accuracy and  the val_accuracy
    axes = ax[1]
    axes.plot(x, history.history['accuracy'], marker='o', label='accuracy')
    axes.plot(x, history.history['val_accuracy'], marker='o', label='val_accuracy')
    axes.set_title('batch_size=' + str(parameters['batch_size'][i]),fontsize=12)
    axes.set_xlabel('epochs',fontsize=12)
    axes.set_ylabel('accuracy',fontsize=12)
    axes.legend(fontsize=12)
    axes.grid()

    i += 1
````

#### 2.3.  Evaluate the model, confusion matrix and classification report.

````python
# best parameters: epoches = 6  and batch_size = 64
epochs = 6 
batch_size = 64

# create cnn model
model = create_cnn()

# training cnn model
model.compile(optimizer='adam',loss='categorical_crossentropy', metrics=['accuracy'])
history = model.fit(
        x_train, y_train, epochs=epochs, batch_size=batch_size , validation_data=(x_test, y_test))
````

````python
fig,ax = plt.subplots(1,2,figsize=(18,9))

x = np.linspace(1,epochs,epochs)
# Plot out the loss and the val_loss
axes = ax[0]
axes.plot(x, history.history['loss'], marker='o', label='loss',)
axes.plot(x, history.history['val_loss'], marker='o', label='val_loss')
axes.set_title('batch_size=' + str(batch_size),fontsize=12)
axes.set_xlabel('epochs',fontsize=12)
axes.set_ylabel('loss',fontsize=12)
axes.legend(fontsize=12)
axes.grid()

# Plot out the accuracy and  the val_accuracy
axes = ax[1]
axes.plot(x, history.history['accuracy'], marker='o', label='accuracy')
axes.plot(x, history.history['val_accuracy'], marker='o', label='val_accuracy')
axes.set_title('batch_size=' + str(batch_size), fontsize=12)
axes.set_xlabel('epochs',fontsize=12)
axes.set_ylabel('accuracy',fontsize=12)
axes.legend(fontsize=12)
axes.grid()
````

````python
from sklearn.metrics import classification_report

# predict
y_pred = (model.predict(x_test)>0.5).astype(int)

# convert one-hot encodings into integers
y_pred_labels = np.argmax(y_pred,axis=1)
y_test_labels = np.argmax(y_test,axis=1)

# print report
print("the classification report:")
print(classification_report(y_test_labels,y_pred_labels))
````

````python
import seaborn as sns
from sklearn.metrics import confusion_matrix

# Calculate confusion matrix
cm = confusion_matrix(y_test_labels,y_pred_labels)

# plot the confusion matrix
plt.figure(figsize=(15,12))
disp = sns.heatmap(cm,annot=True,fmt='', linewidths=.5,cmap='ocean')
disp.set_xlabel("Predicted label",fontsize=13) 
disp.set_ylabel("True label",fontsize=13) 
disp.set_title('Confusion Matrix',fontsize=16)
plt.show()
````
